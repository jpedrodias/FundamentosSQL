r) Eliminar os clientes residentes em Portugal e sem qualquer compra

````python
%%sql
/* Produtos menos comprados por cada cliente (quantidade total mínima) */
SELECT  t.nome_cliente,
        t.nome_produto,
        t.quantidade_total
FROM (
        /* Total comprado de cada produto por cada cliente */
        SELECT  c.id    AS cliente_id,
                c.nome  AS nome_cliente,
                p.nome  AS nome_produto,
                SUM(v.quantidade) AS quantidade_total
        FROM    vendas   v
        JOIN    clientes c ON c.id = v.cliente_id
        JOIN    produtos p ON p.id = v.produto_id
        GROUP BY c.id, p.id
) AS t
JOIN (
        /* Quantidade mínima por cliente (sem WITH) */
        SELECT  s.cliente_id,
                MIN(s.prod_sum) AS min_qtd
        FROM (
                SELECT  v.cliente_id,
                        v.produto_id,
                        SUM(v.quantidade) AS prod_sum
                FROM    vendas v
                GROUP BY v.cliente_id, v.produto_id
        ) AS s
        GROUP BY s.cliente_id
) AS m
  ON  m.cliente_id = t.cliente_id
  AND m.min_qtd    = t.quantidade_total
ORDER BY t.nome_cliente;
````

**Explicação da célula anterior:**
Esta consulta identifica clientes que ainda não fizeram compras. Utiliza LEFT JOIN entre clientes e vendas e WHERE para filtrar clientes sem vendas correspondentes, identificando clientes inativos ou potenciais prospects.

````python
%%sql
/* 1) Total de unidades compradas de cada produto por cada cliente */
WITH total_cliente_produto AS (
    SELECT  c.id                          AS cliente_id,
            c.nome                        AS nome_cliente,
            p.id                          AS produto_id,
            p.nome                        AS nome_produto,
            SUM(v.quantidade)             AS quantidade_total
    FROM    vendas   v
    JOIN    clientes c ON c.id = v.cliente_id
    JOIN    produtos p ON p.id = v.produto_id
    GROUP BY c.id, p.id
),

/* 2) Quantidade mínima comprada por cada cliente */
min_por_cliente AS (
    SELECT  cliente_id,
            MIN(quantidade_total) AS min_qtd
    FROM    total_cliente_produto
    GROUP BY cliente_id
)

/* 3) Produto(s) que têm essa quantidade mínima */
SELECT  t.nome_cliente,
        t.nome_produto,
        t.quantidade_total
FROM    total_cliente_produto t
JOIN    min_por_cliente   m
      ON m.cliente_id = t.cliente_id
     AND m.min_qtd    = t.quantidade_total
ORDER BY t.nome_cliente;
````

**Explicação da célula anterior:**
Esta consulta lista os 5 produtos mais vendidos em termos de quantidade. Utiliza JOIN, GROUP BY para agrupar por produto, SUM para calcular total vendido, ORDER BY DESC para ordenar decrescentemente e LIMIT para mostrar apenas os top 5.

````python
%%sql
SELECT nome_cliente,
       nome_produto,
       quantidade_total
FROM (
    SELECT  c.nome  AS nome_cliente,
            p.nome  AS nome_produto,
            SUM(v.quantidade) AS quantidade_total,
            RANK() OVER (PARTITION BY c.id ORDER BY SUM(v.quantidade) ASC) AS ranking
    FROM    vendas   v
    JOIN    clientes c ON c.id = v.cliente_id
    JOIN    produtos p ON p.id = v.produto_id
    GROUP BY c.id, c.nome, p.nome
) sub
WHERE ranking = 1
ORDER BY nome_cliente;
````

**Explicação da célula anterior:**
Esta consulta lista os 3 clientes que mais gastaram dinheiro. Utiliza múltiplos JOINs, GROUP BY para agrupar por cliente, SUM para calcular valor total gasto, ORDER BY DESC para ordenar decrescentemente e LIMIT para mostrar apenas os top 3.

u) Listar o nome dos fornecedores cujos produtos já venderam mais de 100 unidades e que tiveram mais de 10 vendas

````python
%%sql
SELECT  t.nome_cliente,
        t.nome_categoria,
        t.quantidade_total
FROM (
    -- Quantidade total comprada por cliente e por categoria
    SELECT  c.id       AS cliente_id,
            c.nome     AS nome_cliente,
            cat.nome   AS nome_categoria,
            SUM(v.quantidade) AS quantidade_total
    FROM    vendas       v
    JOIN    clientes     c   ON c.id = v.cliente_id
    JOIN    produtos     p   ON p.id = v.produto_id
    JOIN    categorias   cat ON cat.id = p.id_categoria
    GROUP BY c.id, cat.id
) AS t
JOIN (
    -- Quantidade máxima por cliente
    SELECT  cliente_id,
            MAX(quantidade_total) AS max_qtd
    FROM (
        SELECT  v.cliente_id,
                p.id_categoria,
                SUM(v.quantidade) AS quantidade_total
        FROM    vendas v
        JOIN    produtos p ON p.id = v.produto_id
        GROUP BY v.cliente_id, p.id_categoria
    ) AS resumo
    GROUP BY cliente_id
) AS m
  ON  m.cliente_id = t.cliente_id
  AND m.max_qtd    = t.quantidade_total
ORDER BY t.nome_cliente;
````

**Explicação da célula anterior:**
Esta consulta complexa identifica a categoria de produtos mais comprada por cada cliente. Utiliza subconsultas aninhadas (CTEs) para calcular totais por cliente/categoria e depois identificar a categoria com maior quantidade para cada cliente, mostrando a preferência de compra de cada cliente.
